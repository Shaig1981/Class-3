```python
import yfinance as yf
from matplotlib import pyplot as plt
```

```python
nvidia = yf.download("NVDA", start= "2026-01-01", end= "2026-03-01", interval= "1d")["Close"]
```

```python
nvidia
```

```python
plt.plot(nvidia)
plt.xticks(rotation = 90)
plt.show()
```
